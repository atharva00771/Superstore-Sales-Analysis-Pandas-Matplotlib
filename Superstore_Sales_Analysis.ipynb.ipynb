```python
#category wise sales

category_sales = df.groupby("Category")["Sales"].sum()

plt.barh(category_sales.index, category_sales.values)
plt.title("Category wise sales")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.grid()
plt.savefig('images2/category_sales',dpi=300,bbox_inches='tight')
plt.show()
```

```python
#category wise profit

category_profit=df.groupby("Category")['Profit'].sum()

plt.bar(category_profit.index,
       category_profit.values)

plt.title("categorywise profit")
plt.xlabel("category")
plt.ylabel('Profit')
plt.grid(color='gray')
plt.savefig('images2/category_profit',dpi=300,bbox_inches='tight')
plt.show()
```

```python
#Month wise Sales

months=df.groupby("Order Date")['Sales'].sum().head().sort_values()

plt.plot(months.index,
        months.values,
        marker='^',
        color='black')

plt.title("month wise sales")
plt.xlabel("months")
plt.ylabel("sales")
plt.savefig('images2/month_by_sales',dpi=300,bbox_inches='tight')
plt.show()
```

```python
#category share

category_sales=df.groupby("Category")['Sales'].sum()

plt.pie( category_sales.values,
      labels= category_sales.index,
       autopct='%1.1f%%')
plt.savefig("images2/category_piechart.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
#Quantity Distribustion
plt.hist(df['Quantity'])
plt.title('Quantity_of_Product')
plt.savefig("images2/profit_histogram.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
#Sales vs Profit
plt.scatter(df['Sales'],df['Profit'],label='Sales vs Profit',color='black')
plt.title('sales vs profit')
plt.xlabel('sales')
plt.ylabel('profit')
plt.legend()
plt.savefig("images2/sales_vs_profit",dpi=300,bbox_inches='tight')
plt.show()
    
```
